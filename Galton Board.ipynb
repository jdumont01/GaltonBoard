# Agile Approach

This project will be defined and developed via the Agilet process:
- Create User Stories with Acceptance Criteria
- Assign points to each story
- Order the stories based on priority to create MVP
- Develop and Demonstrate the Deliverables for each Story

## User Stories And Acceptance Criteria

### User Interface

**Story 1:**  As a new user, I want to define the depth of the Galton Board, so that I can change the number of levels of the Galton Board.

**Acceptance Criteria:**  
[ ] The user will have an option to define the depth of the Galton Board.   

**Story 2:**  As a new user, I want to define the number of times the Galton Board is traversed, so that I can analyze how the final distribution of results changes with the number of traversals.

**Acceptance Criteria:**   
[ ] The user will have an option to define the number of time the Galton Board is traversed - called events.   
[ ] The user will have an option to define how quickly each level of the Galton Board is reached - called rate.

**Story 3:**  As a new user, I want to view an extensive set of results, so that I can view the distribution of the each completed event. 

**Acceptance Criteria:**   
[ ] The results will be tallied in real-time, after each event completes.   
[ ] The accumulated number of times an event results in each bucket will be displayed for the user.   
[ ] The fraction of each accumulated bucket divided by the total number of events will be displayed for the user.     
[ ] The expected fraction of each bucket will be calculated and displayed for the user.   
[ ] The traversed pathway from each event will be displayed for the user.

**Story 4:**  As a new user, I want to select large values for the depth of the Galton Board and a large number of events, so that I can create and analyze complex distributions with or without a UI.

**Acceptance Criteria:**   
[ ] The user will be able to define board depths of up to 100.   
[ ] The user will be able to define the total number of events to process up to 1 billion.   
[ ] The user will be able to view raw stats in the UI.   

**Story 5:**  As a user interested in statistics, I want to overlay different common distributions, so that I can compare the results generated by the Galton Board to common distrubitions like:

* Student T-distribution
* Poisson distribution 
* Normal distribution

**Acceptance Criteria:**   
[ ] The user will have the ability to select different "common" distribuions.   
[ ] The user will see an overlay of the distribution and the results of the Galton Board events in a histogram.   

**Story 6:**  As a user interested in studying statistical bias, I want to be able to define a rotational angle, so that the events can be skewed either to the left or the right.

**Acceptance Criteria:**   
[ ] The user will have the ability to select an angle that rotates the board.   
[ ] The board will be displayed with the rotation.

### Historical Results

**Story 7:**  As a user interested in comparing long-term result sets, I want the results of each run to be saved, so that I can recall the results.

**Acceptance Criteria:**  
[ ] The user will have access to the raw data to query.   
[ ] The data will be saved so that each run is uniqely identified.

## Defining Minimal Viable Product (MVP)

There are several stories to review; however, to delvier a minimal viable product (MVP), a first set of deliverables has to be selected.

### MVP Criteria (First Set of Deliverables)

The MVP will be based on the highest value work.  First, the basic product needs to work:
- how the Galton Board will look based on user input.
- how many times it will be traversed based on user input.
- how the output of the results will be presented to the user. 

### Second Set of Deliverables

The second set of deliverables will allow the user to perform statistical comparisons.  This includes the following features:
- overlaying a user-defined distribution on the results.
- retrieving historical data

### Third Set of Deliverables

The third set of deliverables will allow the user to query historical data.  

### Fourth Set of Deliverables

The fourth set of deliverables will be based on different ways to use a Galton Board to peform statistical bias run by rotating the Galton Board.
